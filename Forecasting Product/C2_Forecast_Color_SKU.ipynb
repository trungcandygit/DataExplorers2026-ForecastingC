```python
# thư viện
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pmdarima as pm
```

```python
# tải dữ liệu lịch sử và tương lai
# YYYY-MM-DD (ngày đầu tháng)
df        = pd.read_csv('../tnbike_data.csv',        index_col=0, parse_dates=True)
future_df = pd.read_csv('../future.csv',             index_col=0, parse_dates=True)
future_df.head()
```

```python
# đổi tên biến
df = df.rename(columns={'revenue': 'y'})
df.head(1)
```

```python
# trích xuất biến phụ (regressors)
train_X  = df.iloc[:, 1:]
future_X = future_df.iloc[:, 1:]
future_X.head()
```

# Mô hình SARIMAX

```python
# lấy tham số tốt nhất
parameters = pd.read_csv("../Forecasting Product/best_params_sarimax.csv", index_col=0)
parameters
```

```python
# lưu từng tham số riêng lẻ
p = int(parameters.loc["p"][0])
d = int(parameters.loc["d"][0])
q = int(parameters.loc["q"][0])
P = int(parameters.loc["P"][0])
D = int(parameters.loc["D"][0])
Q = int(parameters.loc["Q"][0])
```

```python
# mô hình
# tần suất tháng → seasonal = 12
model = pm.ARIMA(order=(p, d, q),
                 seasonal_order=(P, D, Q, 12),
                 suppress_warnings=True,
                 enforce_stationarity=False)
model.fit(df.y)
```

# Dự báo

```python
# dự báo Q2/2026 (T4, T5, T6)
predictions_sarimax = model.predict(n_periods=len(future_df),
                                    X=future_X).rename("sarimax")
```

```python
# trực quan hóa
df['y']['2025-01-01':].plot(figsize=(15, 6), legend=True);
predictions_sarimax.plot(legend=True);
```

```python
# xuất kết quả
predictions_sarimax.to_csv("Ensemble/predictions_sarimax.csv")
```
